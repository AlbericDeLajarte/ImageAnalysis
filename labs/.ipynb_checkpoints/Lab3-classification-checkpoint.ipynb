##### Analytical expression
We compute $g_{ij}$ between all the classes. Then, for a given point $x$ we compute where it belongs. We proceed by elemination as in this example :
- A or B -> $x\notin B$
- A or C -> $x\notin C$
- B or C -> $x\notin C$

If a point doesnt not belong to B and C, we conclue it's in A.

From the dataset visualization, we see that all the classes have almost diagonal covariance matrix. We can thus compare two bayesian models, one assuming a diagonal covariance matrix, and the other one with a complete covariance matrix for reference.

```python
# Analytical expression of Bayesian discriminant function with covariance matrix and no constant 
def discriminant_bayesian_complete(mu, var, XX, YY):
    def foo(X, mu, var):
        mu = np.asmatrix(mu).T
        var = np.asmatrix(var)
        g = -0.5*(X.T@var.I@X) + 0.5*X.T@var.I@mu - 0.5*mu.T@var.I@mu + 0.5*mu.T@var.I@X - np.log(2*np.pi*np.sqrt(LA.norm(var)))
        return g
    G = np.zeros((len(YY[:,0]), len(XX[0,:])))
    for i, x in enumerate(XX[0,:]):
        for j, y in enumerate(YY[:,0]):
            X = np.matrix([[x], [y]])
            #print(X)
            G[j,i] = foo(X, mu, var)
    return G
```

```python
# Analytical expression of Bayesian discriminant function with diagonal covariance matrix and no constant 
def discriminant_bayesian_diagonal(mu, var, XX, YY):
    C = np.transpose(mu)@LA.inv(var)@mu/(-2)
    G = (var[1,1]*XX**2 + var[0,0]*YY**2)/(-2*var[1,1]*var[0,0]) + (var[1,1]*XX*mu[0] + var[0,0]*YY*mu[1])/(var[1,1]*var[0,0]) + C
    return G
```

```python
# Create 2D space
xmin, xmax = (-20, 16)
ymin, ymax = (-8, 8)
nPoint = 200
XX, YY = np.meshgrid(np.linspace(xmin, xmax, nPoint), np.linspace(ymin, ymax, nPoint))
#Create dictionnaries that will contains information related to each class
G_a = {"name":"G_a",
       "value_diagonal":discriminant_bayesian_diagonal(mu_a, var_a, XX, YY),
       "value_complete":discriminant_bayesian_complete(mu_a, var_a, XX, YY),
       "var":(var_a),
       "mu":(mu_a),
       "datapoint":a,
       "marker":"+",
       "color":'r',
      "indice": 1}
G_b = {"name":"G_b",
       "value_diagonal":discriminant_bayesian_diagonal(mu_b, var_b, XX, YY),
       "value_complete":discriminant_bayesian_complete(mu_b, var_a, XX, YY),
       "var":(var_b),
       "mu":(mu_b),
       "datapoint":b,
       "marker":"o",
       "color":"g",
      "indice": 2}
G_c = {"name":"G_c",
       "value_diagonal":discriminant_bayesian_diagonal(mu_c, var_c, XX, YY),
       "value_complete":discriminant_bayesian_complete(mu_c, var_c, XX, YY),
       "var":(var_c),
       "mu":(mu_c),
       "datapoint":c,
       "marker":",",
       "color":"b",
      "indice": 3}
Gs = [G_a, G_b, G_c]

#Compute the zone that belongs to class
for v, b in zip(["value_diagonal", "value_complete"], ["bool_zone_diagonal", "bool_zone_complete"]):
    G_a_b = G_a[v]-G_b[v]>0
    G_a_c = G_a[v]-G_c[v]>0
    G_b_c = G_b[v]-G_c[v]>0
    G_a[b] = np.logical_and(G_a_b, G_a_c)
    G_b[b] = np.logical_and(np.logical_not(G_a_b), G_b_c)
    G_c[b] = np.logical_and(np.logical_not(G_a_c), np.logical_not(G_b_c))
```

```python
# Plot the function with the dataset
fig, axes = plt.subplots(4, len(Gs), figsize = (15, 20), sharey=True, sharex=True)
for i, v in enumerate(["value_diagonal", "value_complete", "bool_zone_diagonal", "bool_zone_complete"]):
    for j, (G, axe) in enumerate(zip(Gs, axes[i,:])):
        if j == 0: axe.set_ylabel(v)
        axe.pcolor(XX,YY, G[v])
        axe.scatter(G["datapoint"][:,0], G["datapoint"][:,1], marker = G["marker"])
        if i == 0: axe.set_title(G["name"])
plt.show()
```

**Analysis** : On the plots above, we compared a simplified version that contains only diagonal terms in the covariance matrix to the full covariant matrix.  
The zone in yellow correspond to the zoning to to corresponding class. We see few differences between the *full* and the *diagonal* version.  
The zoning gives a better view of the **contour** that separate classes.

Note that left from class `c`, we see that the probabilities are higher to be in classes `a` and `b`.

### 1.2 Mahalanobis distance
For classes `a` and `b`, give the expression of the Mahalanobis distance used to classify a point in class `a` or `b`, and verify the obtained classification, in comparison with the "complete" Bayes classification, for a few points of the plane.

*Add your implementation and discussion*

We use a covariance matrix of own $\mu$'s of each class. The Mahalanobis distance is computed from the `scipy` library.
To be easily comparable with probability method, we store the **opposite** of the Mahalanobis distance.
To assign a datapoint to a class, we compute the distance to the 3 $\mu$ and select the **longest** one.

```python
from scipy.spatial.distance import mahalanobis
for v in (Gs): #Gs is a list of dictionaries. Each dictionary contains datas of a class (a, b or c)
    v["value_mahalanobis"] = np.asarray([[-mahalanobis(np.asarray([x,y]), v["mu"], LA.inv(v["var"])) for x in (XX[0,:])] for y in (YY[:,0])])
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 5))
for ax, G in zip(axes, Gs[0:3]):
    ax.pcolor(XX,YY, G["value_mahalanobis"])
    ax.scatter(G["datapoint"][:,0], G["datapoint"][:,1], marker = G["marker"], color = G["color"])
    ax.set_title("class :{}, mu is at ({:.3f},{:.3f})".format(G["name"][-1],*G["mu"]))
plt.show()
```

This is a the distance map with the opposite Mahalanobis distance.  
Let's look at the zoning of `a` and `b`.

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 7.5))
G_a_b_mahalanobis = G_a["value_mahalanobis"]-G_b["value_mahalanobis"]>0 #Smallest distance -> in class a
G_a_b_bayse = G_a["value_complete"]-G_b["value_complete"]>0 #Biggest probability -> in class a
for ax, v, t in zip(axes, [G_a_b_mahalanobis, G_a_b_bayse], ["Mahalanobis distance classifier", "Baysian classifier"]):
    ax.pcolor(XX,YY, v)
    ax.scatter(G_a["datapoint"][:,0], G_a["datapoint"][:,1], marker = G_a["marker"], color = G_a["color"])
    ax.set_title(t)
    ax.scatter(G_b["datapoint"][:,0], G_b["datapoint"][:,1], marker = G_b["marker"], color = G_b["color"])
plt.show()
```

**Analysis** The zoning is very similar between both methods. Few differences could come from the constant term $\log{2\pi \sqrt{|V|}}$ that is absent from Mahalanobis distance. It could also come from the way the distance is unilaterally computed as 
$$
x^T\cdot V^{-1}\cdot \mu \neq 0.5\cdot x^T V^{-1}\mu + 0.5\cdot \mu^T V^{-1} x
$$

### 1.4 Comparison
To have an overview of the different methods of classification, we created a final plot that shows the zoning of each class by different methods as well as measuring their relative quality by classing the dataset points.

```python
def segment_class(size_x, size_y, G_a, G_b, G_c):
    G_seg = np.zeros(shape=(size_x, size_y))
    for i in range(size_x):
            for j in range(size_y):
                if G_a[i, j] > G_b[i, j]:
                    if G_a[i, j]> G_c[i, j]:
                        G_seg[i, j] = 1
                    else: 
                        G_seg[i, j] = 3
                else:
                    if G_b[i, j]> G_c[i, j]:
                        G_seg[i, j] = 2
                    else: 
                        G_seg[i, j] = 3
    return G_seg
```

```python
string_method = ["complete", "diagonal", "mahalanobis", "euclidean"]
G_seg = dict(zip(string_method, [segment_class(len(XX), len(YY), G_a["value_"+s], G_b["value_"+s], G_c["value_"+s]) for s in string_method]))

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
for row, axe in enumerate(axes):
    for ax, s in zip(axe, string_method[2*row:2*row+2]):
        ax.pcolor(XX, YY, G_seg[s])
        for G in Gs:
            ax.scatter(G["datapoint"][:,0], G["datapoint"][:,1], marker = G["marker"], color = G["color"])
            ax.scatter(G["mu"][0], G["mu"][1], marker = G["marker"], color = "black")

        ax.set_title("Method : {}".format(s))
```

**Analysis** : The `Complete`, `Diagonal` and `Mahalanobis` gives similar results. But they require to have a gaussian distribution in order to perform well.

```python
def find_class(XY, G):
    indiceY = nPoint/(xmax-xmin) *(XY[:,0]-xmin)
    indiceX = nPoint/(ymax-ymin) *(XY[:,1]-ymin)
    
    indiceX = indiceX.astype(int)
    indiceY = indiceY.astype(int)
    
    return G[indiceX, indiceY]
```

```python
for g in Gs:
    print("Dataset: {}".format(g["name"]))
    for s in string_method:
        data_classified = find_class(g["datapoint"], G_seg[s])
        print("Succesful classification using {} method is {} %".format(s, 100*len(data_classified[data_classified==g["indice"]])/len(data_classified)))

    print("\n")
    
    
```

**Analysis** : Classing the points in the dataset back to their classes gives the best score to Mahalanobis. It's good to remember that with such a small number of point, the performence of the complete baysian method is not enlighted.

### 2.1 Dataset loading
Here we first declare the methods `extract_data` and `extract_labels` so that we can reuse them later in the code.
Then we extract both the data and corresponding labels, and plot randomly some images and corresponding labels of the training set.

```python
import gzip
import numpy as np
import matplotlib.pyplot as plt

def extract_data(filename, image_shape, image_number):
    with gzip.open(filename) as bytestream:
        bytestream.read(16)
        buf = bytestream.read(np.prod(image_shape) * image_number)
        data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
        data = data.reshape(image_number, image_shape[0], image_shape[1])
    return data


def extract_labels(filename, image_number):
    with gzip.open(filename) as bytestream:
        bytestream.read(8)
        buf = bytestream.read(1 * image_number)
        labels = np.frombuffer(buf, dtype=np.uint8).astype(np.int64)
    return labels
```

```python
image_shape = (28, 28)
train_set_size = 60000
test_set_size = 10000

data_part2_folder = os.path.join(data_base_path, data_folder, 'part2')

train_images_path = os.path.join(data_part2_folder, 'train-images-idx3-ubyte.gz')
train_labels_path = os.path.join(data_part2_folder, 'train-labels-idx1-ubyte.gz')
test_images_path = os.path.join(data_part2_folder, 't10k-images-idx3-ubyte.gz')
test_labels_path = os.path.join(data_part2_folder, 't10k-labels-idx1-ubyte.gz')

train_images = extract_data(train_images_path, image_shape, train_set_size)
test_images = extract_data(test_images_path, image_shape, test_set_size)
train_labels = extract_labels(train_labels_path, train_set_size)
test_labels = extract_labels(test_labels_path, test_set_size)
```

```python
prng = np.random.RandomState(seed=123456789)  # seed to always re-draw the same distribution
plt_ind = prng.randint(low=0, high=train_set_size, size=10)

fig, axes = plt.subplots(1, 10, figsize=(12, 3))
for ax, im, lb in zip(axes, train_images[plt_ind], train_labels[plt_ind]):
    ax.imshow(im, cmap='gray')
    ax.axis('off')
    ax.set_title(lb)
```

### 2.2 MLP
*Add your implementation and discussion*

```python
h = 100
train_images_flat = train_images.reshape(train_set_size, -1)
test_images_flat = test_images.reshape(test_set_size, -1)
print("Training: {}, Test: {}".format(train_images_flat.shape, test_images_flat.shape))
```

```python
# Accuracy
def accuracy(key, solver):
    predicted = np.argmax(solver.predict_proba(test_images_flat), 1)
    score =  sum(predicted == test_labels) / test_set_size
    print("Our score for {} {:.2%} of correct answers".format(key, score))
```

```python
# Train 
from tqdm import tqdm
from sklearn.neural_network import MLPClassifier
mlp_lbfgs = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(h,))
mlp_sgd = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(h,))
mlp_adam = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes= (h,))
mlps = {"lbfgs":mlp_lbfgs, "sgd":mlp_sgd, "adam":mlp_adam}
for key in tqdm(mlps):
    mlps[key].fit(train_images_flat, train_labels)
```

```python
for key in mlps:
    accuracy(key, mlps[key])
```

We can now compare the performance of these different methods to select the most suited one to our problem

```python
for alpha_rate in np.logspace(-5, 1, 7):
    
    mlp_lbfgs = MLPClassifier(solver='lbfgs', alpha=alpha_rate, hidden_layer_sizes=(h,))
    mlp_sgd = MLPClassifier(solver='sgd', alpha=alpha_rate, hidden_layer_sizes=(h,))
    mlp_adam = MLPClassifier(solver='adam', alpha=alpha_rate, hidden_layer_sizes= (h,))
 
    mlps = {"lbfgs":mlp_lbfgs, "sgd":mlp_sgd, "adam":mlp_adam}
    
    print("Using alpha = {}".format(alpha_rate))
    for key in mlps:
        mlps[key].fit(train_images_flat, train_labels)    
        accuracy(key, mlps[key])
    
    print("\n")
```

We see that even with low alpha value, the "lbfgs" method doesn't give good result compare to the other two methods. Even though it is the fastest of the three methods, we will now only use the default "adam" method which is described as better for large dataset.

A very important point to consider in machine learning problem is overfitting. One way to check for that is to compare the performance of the train data to the test data for prediction. 

```python
for alpha_rate in np.logspace(-6, 2, 9):
    print("Using alpha = {}".format(alpha_rate))
    
    mlp_adam = MLPClassifier(solver='adam', alpha=alpha_rate, hidden_layer_sizes= (h,))
    mlp_adam.fit(train_images_flat, train_labels)
    
    predicted_test = np.argmax(mlp_adam.predict_proba(test_images_flat), 1)
    score_test =  sum(predicted_test == test_labels) / test_set_size
    
    predicted_train = np.argmax(mlp_adam.predict_proba(train_images_flat), 1)
    score_train =  sum(predicted_train == train_labels) / train_set_size
    
    print(score_train/score_test)
    
    print("\n")
```

We see here that for high value of alpha, the ratio gets closer to one, which is a good sign that we have reduced overfitting with the penalty term alpha. As we also know that for alpha between 1e-5 and 10, the adam method gives very similar performances, we can conclude that a value around 10 would be a good choice for alpha.
